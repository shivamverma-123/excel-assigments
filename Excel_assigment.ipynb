Q1. Explain the difference between Relative , Absolute and mixed cell references?

Ans: The difference between Relative , Absolute and mixed cell references are

 1. Relative Cell Reference
 Definition: Changes automatically when the formula is copied to another cell.
 Symbol: No dollar sign ($).
 Example: =A1 + B1

2. Absolute Cell Reference

 Definition: Does not change when you copy the formula to another cell.
 Symbol: Dollar signs before both column and row → $A$1
 Example: =$A$1 + B1

3. Mixed Cell Reference

 Definition: Only one part (either row or column) stays fixed, the other changes.
 Symbol: One dollar sign ($).
 Examples:
o $A1 → Column A is fixed, row changes
o A$1 → Row 1 is fixed, column changes

Q7.Create a column chart showing total sales by product.

ANS: To create a column chart showing the total sales by product is
      Step 1: Add a “Total Sales” column
              In a new column (say, G2), type this formula:
               =E2*F2
               drag it down f or all rows.
               you’ll have total sales for each record.
      Step 2: Summarize Total Sales by Product
              1.Select your dataset (including header)
              2.Go to the Insert tab
              3.Click PivotTable → choose “New Worksheet.”
              4.In the PivotTable Field List:
                  Drag Product → to Rows area.
                  Drag Total Sales → to Values area (it will show Sum of Total Sales).
             Now you’ll see total sales per product.
       Step 3: Create the Column Chart
              1. Click anywhere inside the PivotTable.
              2. Go to Insert → choose Column Chart → Clustered Column.
              3. You’ll get a chart showing Product on the X-axis and Total Sales on the Y-axis.
